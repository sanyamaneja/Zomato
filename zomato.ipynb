```python
# Importing 2nd dataset
b=pd.read_excel("C:\\Users\\Sanyam Aneja\\Downloads\\Zomato_Country-Code.xlsx")
b
```

```python
# merging both datasets based on common column "Country Code"
c=pd.merge(a,b,on='Country Code',how='left')
c
```

```python
# checking null values of new table with heatmap
sns.heatmap(c.isnull());
```

```python
# How many Countries are there in this DataSet?
c['Country'].value_counts()
```

```python
plt.bar(c['Country'],c['Price range'])
plt.xticks(rotation=90)
plt.show()
```

```python
country_names=c['Country'].value_counts().index
country_names
```

```python
country_val=c['Country'].value_counts().values
country_val
```

```python
# Which country received highest orders?
plt.scatter(country_names,country_val);
plt.xticks(rotation=90,fontsize=14);
```
